```python
def is_hot(temp):
    if temp > 25:
        print(temp)
for i in df.Temperature:
    is_hot(i)
```

```python
counter = 0 
for i in df.Temperature:
    if i >25:
        counter+=1
print('The number of NYC hot days is:', counter)
```

```python
rainy_days, snowy_days = 0,0
for i in df.Events:
    if i == 'Rain':
        rainy_days +=1
    elif i == 'Snow':
        snowy_days+=1

print(f"NYC has {rainy_days} rainy days and {snowy_days} snowy days")
```
